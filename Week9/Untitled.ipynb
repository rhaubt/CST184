```r
# Create binomial distribution data
n <- 10     # number of trials
p <- 0.6    # probability of success

binom_df <- data.frame(
  successes = 0:n,
  probability = dbinom(0:n, size = n, prob = p)
)

# Plot the distribution
ggplot(binom_df, aes(x = successes, y = probability)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(
    title = "Binomial Distribution (n = 10, p = 0.6)",
    x = "Number of Successes",
    y = "Probability"
  ) +
  theme_minimal()
```

```r
# Create Poisson distribution data
lambda <- 4

poisson_df <- data.frame(
  events = 0:15,
  probability = dpois(0:15, lambda = lambda)
)

# Plot the distribution
ggplot(poisson_df, aes(x = events, y = probability)) +
  geom_bar(stat = "identity", fill = "firebrick") +
  labs(
    title = "Poisson Distribution ( = 4)",
    x = "Number of Events",
    y = "Probability"
  ) +
  theme_minimal()
```
